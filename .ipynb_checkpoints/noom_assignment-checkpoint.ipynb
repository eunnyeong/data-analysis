```python
import pandas as pd
```

```python
raw_data = pd.read_csv("~/Downloads/noom/noom_user.csv", parse_dates=["Purchased At"])

print(raw_data.columns)

print(raw_data.shape)

raw_data.head()
```

1.

```python
# 이 컬럼을 제외한 나머지 컬럼을 제거

columns = ['Access Code', 'Name', 'Gender', 'Age', 'Height', 
           'Initial Weight', 'Lowest Weight', 'Target Weight', 
           'Status', 'Price', 'Purchased At', 'Channel']

data = raw_data[columns]

data = data.set_index('Access Code')

data.head()
```

```python
# 성별(Gender) 컬럼에서 중복된 값을 제거합니다.
# 최종적으로 성별 컬럼의 종류가 나옵니다. (nan은 Not a Number, 즉 값이 없다는 의미입니다)
raw_data["Gender"].unique()
```

```python
# 대문자로 되어있는 텍스트를 소문자로 변경

data["Gender(clean)"] = data["Gender"].replace("MALE", "male").replace("FEMALE", "female")

data[["Name", "Gender", "Gender(clean)"]].head()
```

```python
# 전체 가입자중 남성의 비율과 여성의 비율을 계산

data['Gender(clean)'].value_counts()
```

```python
# Height 컬럼에서 가장 작은(min) 값을 찾아와서 출력합니다.
raw_data["Height"].min()
```

```python
import numpy as np

# 키가 -1 cm인 사람은 NaN으로 데이터를 넣어주세요.
# data["Height(clean)"] = data["Height"].replace(-1, np.nan)

data["Height(clean)"] = data["Height"]
data.loc[data["Height"] == -1, "Height(clean)"] = np.nan

data.loc[data["Height"] == -1, ["Name", "Height", "Height(clean)"]].head()
```

```python
# 1) 최소/최대/평균 키를 구하고, 

print(data['Height(clean)'].min())
print(data['Height(clean)'].max())
print(data['Height(clean)'].mean())
```

```python
# 2) 남성/여성별 평균 키를 구해주세요.

pd.pivot_table(data, index = 'Gender(clean)', values = ['Height(clean)'])
```
