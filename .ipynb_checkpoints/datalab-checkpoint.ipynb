어제 쇼핑몰에서 기모팬츠를 50% 세일한다는 알람이 왔다.
- 날씨의 변화 때문인지, 데이터의 변화가 있는지 궁금해졌다.
- 네이버 데이터랩의 데이터로 1년 동안의 데이터를 받아서 분석했다.
- 따뜻한 날씨를 나타내는 '원피스'와 '기모팬츠'를 비교하기 위해 두 가지 데이터를 추출했다.

```python
import pandas as pd 
```

```python
data = pd.read_excel("./datalab.xlsx")
data.head()
```

```python
# 상위 6개의 Data 는 NaN 이 있어서 삭제

data = pd.read_excel("./datalab.xlsx", header=6)
data.head()
```

```python
# 한 개의 날짜 데이터만 남기기

data = data.drop('날짜.1', axis=1)
data.head()
```

```python
# 날짜 데이터를 datetime으로 형변환

data['날짜'] = pd.to_datetime(data['날짜'])

data.dtypes
```

```python
# 정리한 데이터를 엑셀 파일로 내보내기

data.to_excel('./new_datalab.xlsx')
```

이제 그래프를 그려보자

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = "NanumGothic")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches([14, 10])

# y 축의 범위 지정
ax1.set_ylim(0,25)
ax2.set_ylim(0,110)

sns.lineplot(data = data, x = '날짜', y = '기모팬츠', ax = ax1)
sns.lineplot(data = data, x = '날짜', y = '원피스', ax = ax2)
```

비교하기 위해 하나로 통합

```python
fig, ax1 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches([14, 10])

# y 축의 범위 지정
ax1.set_ylim(0,100)

sns.lineplot(data = data, x = '날짜', y = '기모팬츠', ax = ax1)
sns.lineplot(data = data, x = '날짜', y = '원피스', ax = ax1)

ax1.set_ylabel('')
plt.show()
```

그럼 기모팬츠가 가장 많이 검색된 날은 언제 일까

```python
기모팬츠_최대_조회수_열 = data.sort_values(by=['기모팬츠'], ascending=False)

print(기모팬츠_최대_조회수_열.head(1))
pd.DataFrame(기모팬츠_최대_조회수_열.head(1)['날짜'])
```

원피스, 기모팬츠의 검색 수가 지그재그 여서 추이를 그려보자

pandas 변동 로 검색
다음 URL 참조
  - http://jonathansoma.com/lede/foundations-2018/pandas/rolling-averages-in-pandas/
fig, (ax1) = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches([14, 10])

```python
data.rolling(14, on='날짜').mean().plot(x='날짜', ylim=(0, 100))
```

### 결론

그래프를 보니 기모팬츠는 3월 이후에는 검색량이 없고 원피스는 점점 늘어나는 것을 볼 수 있다.
3월 부터 봄옷의 판매량을 늘어나는 것을 차트로 확인했다.

```python
# 14일 전부터 평균 값
data.rolling(14).mean().sort_values(by=['기모팬츠'], ascending=False).head(5)
```
