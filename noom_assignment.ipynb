```python
# 이 컬럼을 제외한 나머지 컬럼을 제거

new_columns = ['Name', 'Age(clean)', 'Height(clean)',
               'Initial Weight', 'Lowest Weight', 'Target Weight', 'Status']

weight = data[new_columns].copy()

weight.head()
```

```python
# Weight Loss(goal) - 목표 감량치. Initial Weight 컬럼과 Target Weight의 차이를 나타냅니다. (마이너스가 나올 수 있습니다)
# Weight Loss(current) - 최대 감량치. Initial Weight 컬럼과 Lowest Weight의 차이를 나타냅니다. (마이너스가 나올 수 있습니다)
# 체질량지수(BMI) - 키(Height(clean))와 체중(Initial Weight)으로 체지방의 양을 추정하는 공식입니다.

weight["Weight Loss(goal)"] = weight["Initial Weight"] - weight["Target Weight"]
weight["Weight Loss(current)"] = weight["Initial Weight"] - weight["Lowest Weight"]
weight["BMI"] = weight["Initial Weight"] / (weight["Height(clean)"] * weight["Height(clean)"] * 0.0001)

weight[["Name", "Weight Loss(goal)", "Weight Loss(current)", "BMI"]].head()
```

```python
# 필수 (다음의 조건을 만족하지 않으면 Invalid값에는 언제나 False가 들어갑니다)
# 눔의 프로그램을 결제한 구매자. (Status == "completed")

# 옵션 (다음의 조건 중 하나라도 만족할 경우 Invalid값에 True가 들어가야 합니다)
# 나이(Age(clean)), 키(Height(clean))와 몸무게(Initial Weight, Lowest Weight, Target Weight) 중 어느 하나라도 NaN이 들어가 있는 경우.
# 키를 너무 작게 기입했거나(140cm 미만)나, 정 반대로 너무 크게(200cm 초과) 기입한 사용자.
# BMI수치가 너무 낮거나(18.5 미만) 너무 높은 사용자. (30.0 초과)
# 목표 감량치(Weight Loss(goal))가 마이너스인 경우. (보통 현재 체중보다 목표 체중을 낮게 설정합니다)

def find_invalid_user(row):
    if row['Status'] != 'completed':
        return False
    
    if pd.isnull(row['Age(clean)']) or pd.isnull(row['Height(clean)']):
        return True
    
    if pd.isnull(row['Initial Weight']) or pd.isnull(row['Lowest Weight']) or pd.isnull(row['Target Weight']):
        return True
    
    if row['Height(clean)'] < 140 or row['Height(clean)'] > 200:
        return True
    
    if row['BMI'] < 18.5 or row['BMI'] > 30:
        return True
    
    if row['Initial Weight'] < row['Target Weight']:
        return True
    
    else:
        return False

weight['Invalid'] = weight.apply(find_invalid_user, axis=1)
print(weight[weight['Invalid'] == True].shape)
```

```python
Invalid_user = weight[weight['Invalid']]
Invalid_user.head()
```

```python
# 눔의 프로그램을 결제한 구매자. (Status == "completed")
# 목표 감량치(Weight Loss(goal)), 최종 감량치(Weight Loss(current)), BMI 수치 모두 NaN이 아닌 값이 들어가 있는 사용자.
# 최종 감량치(Weight Loss(current))가 10kg 이상.
# BMI 수치가 높은 사용자. (30.0 이상)
# 최종 감량치(Weight Loss(current))가 목표 감량치(Weight Loss(goal))보다 큰 경우. (다이어트에 성공한 사람)

def find_vip_user(row):
    if row['Status'] != 'completed':
        return False
    
    if pd.isnull(row['Weight Loss(goal)']) or pd.isnull(row['Weight Loss(current)']) or pd.isnull(row['BMI']):
        return False
    
    if row['Weight Loss(current)'] < 10:
        return False
    
    if row['BMI'] < 30:
        return False
    
    if row['Weight Loss(current)'] < row ['Weight Loss(goal)']:
        return False
    
    else:
        return True
    
weight['VIP'] = weight.apply(find_vip_user, axis=1)
print(weight[weight['VIP'] == True].shape)
```

```python
# 나이의 경우 다음의 기준으로 그룹화

def group_age(row):
    
    age = row['Age(clean)']
    
    if age <= 17:
        return '00 ~ 17'
    
    elif (age >= 18) & (age <= 24):
        return '18 ~ 24'
    
    elif (age >= 25) & (age <= 35):
        return '25 ~ 35'
    
    elif (age >= 36) & (age <= 44):
        return '36 ~ 44'
    
    elif (age >= 45) & (age <= 54):
        return '45 ~ 54'
    
    else:
        return '55 ~ 99'
    
data['Age(Group)'] = data.apply(group_age, axis=1)
data[['Age(clean)', 'Age(Group)']].head()
```

```python
# 성별과 나이별 결제 / 캔슬 / 환불의 총 인원 수와 비율

table = pd.pivot_table(data, 
                       values='Gender', 
                       index=['Gender(clean)', 'Age(Group)'], 
                       columns='Status', 
                       aggfunc=len, 
                       fill_value=0)

table['total'] = table['cancelled'] + table['completed'] + table['refunded']

table['conversion'] = table['completed'] / table['total'] 

table
```

```python
# 먼저 코칭 데이터를 가져옵니다.
# 이 결과를 coach라는 이름의 변수에 할당합니다.
coach = pd.read_csv("~/Downloads/noom/noom_coach.csv", index_col="Access Code")

# coach 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(coach.shape)

# coach 데이터의 상위 5개를 출력합니다.
coach.head()
```

```python
min_data = data[['Name', 'Status']]

min_data.head()
```

```python
# 기존의 데이터와 코치 데이터를 합쳐주세요.

coach_data = pd.concat([min_data, coach], axis=1)

coach_data.head()
```

```python
# # 코치별 담당 사용자(total) / 구매 완료 횟수(completed) / 캔슬 횟수(canceled) / 환불 횟수(refunded)

# table = pd.pivot_table(coach_data, 
#                        index='Status',
#                        aggfunc='sum', 
#                        fill_value=0)

# table = table.T

# table.head()
```

```python
# 코치별 담당 사용자(total) / 구매 완료 횟수(completed) / 캔슬 횟수(canceled) / 환불 횟수(refunded)

coach_status = coach_data.groupby('Status').sum()

coach_status = coach_status.T

coach_status.head()
```

```python
# 코치별 전환율(conversion rate) / 취소율(cancellation rate)를 계산

coach_status['total'] = coach_status['cancelled'] + coach_status['completed'] + coach_status['refunded']

coach_status['conversion rate'] = coach_status['completed'] / coach_status['total']

coach_status['cancellation rate'] = (coach_status['cancelled'] + coach_status['refunded']) / coach_status['total']

coach_status.head()
```

```python
# 코치별 전환율(conversion rate) 높은 순으로 정렬. 단 모수가 부족한 경우를 배제하기 위해, 코칭을 100회 이상 하지 않은 사용자는 배제

coach_status_top = coach_status[coach_status['total'] >= 100]

coach_status_top.sort_values(by='conversion rate', ascending = False).head()
```

```python
# 코치별 취소율(cancellation rate) 높은 순으로 정렬. 단 모수가 부족한 경우를 배제하기 위해, 코칭을 100회 이상 하지 않은 사용자는 배제

coach_status_top = coach_status[coach_status['total'] >= 100]

coach_status_top.sort_values(by='cancellation rate', ascending = False).head()
```
